```python
# Filter data by parameters in order to analyze separately and copy them to avoid changing original data 
crime_data_filtered_ih = crime_data[
    (crime_data["Series"] == "Intentional homicide rates per 100,000") # Filter by intentional homicide
].copy()
crime_data_filtered_theft = crime_data[
    (crime_data["Series"] == "Theft at the national level, rate per 100,000 population") # Filter by theft
].copy()
crime_data_filtered_assault = crime_data[
    (crime_data["Series"] == "Assault rate per 100,000 population") #Filter by assault
].copy()
crime_data_filtered_kidnapping = crime_data[
    (crime_data["Series"] == "Kidnapping at the national level, rate per 100,000") # Filter by kidnapping
].copy()
crime_data_filtered_sexual_violence = crime_data[
    (crime_data["Series"] == "Total Sexual Violence at the national level, rate per 100,000") # Filter by kidnapping
].copy()
```

```python
# Common data preprocessing function
def preprocess_data(crime_data, education_data):
    # Convert 'Year' to numeric in both dataframes
    crime_data['Year'] = pd.to_numeric(crime_data['Year'], errors='coerce', downcast='float')
    education_data['Year'] = pd.to_numeric(education_data['Year'], errors='coerce', downcast='float')

    # Normalize country names: lowercase and remove extra spaces
    crime_data['Country'] = crime_data['Country'].str.strip().str.lower()
    education_data['Country'] = education_data['Country'].str.strip().str.lower()

    # Merge the two dataframes on 'Country' and 'Year'
    merged_data = pd.merge(
        crime_data,
        education_data,
        on=['Country', 'Year'],
        how='inner'
    )

    # Convert 'Value' and 'Education_Percentage' to numeric
    merged_data['Value'] = pd.to_numeric(merged_data['Value'], errors='coerce', downcast='float')
    merged_data['Education_Percentage'] = pd.to_numeric(merged_data['Education_Percentage'], errors='coerce', downcast='float')

    # Drop rows with missing values in relevant columns
    merged_data = merged_data.dropna(subset=['Value', 'Education_Percentage'])

    return merged_data
```

```python
# Function to plot the relationship
def plot_relationship(merged_data, title, ylabel):
    # Create a scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(
        merged_data['Education_Percentage'],
        merged_data['Value'],
        alpha=0.7,
        color='blue',
        label='Countries'
    )

    # Extract data for trend line calculation
    x = merged_data['Education_Percentage'].values
    y = merged_data['Value'].values

    # Perform linear regression to calculate slope and intercept
    slope, intercept = np.polyfit(x, y, 1)

    # Plot the trend line
    plt.plot(x, slope * x + intercept, color='red', label='Trend Line')

    # Add title and labels
    plt.title(title, fontsize=14)
    plt.xlabel('Education Percentage (%)', fontsize=12)
    plt.ylabel(ylabel, fontsize=12)

    # Add grid and legend
    plt.grid(alpha=0.3)
    plt.legend()

    # Display the plot
    plt.show()
```
